# Exploratory Data Analysis

## Gender

```python
#plotting a bar chart and the count of them

a = sns.countplot(x = 'Gender',data = df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

am_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)

sns.barplot(x= 'Gender',y= 'Amount',data=am_gen)
```

*From the above two graphs , it is clear that the number of female to buy is maximum and their buying power is more.*

## Age

```python
b= sns.countplot(x='Age Group',hue='Gender',data = df)
for bars in b.containers:
    b.bar_label(bars)

# hue is used here to specify the gender in specific age group
```

```python
# plotting a bar graph for age group vs total amount

sm_age= df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x='Age Group',y='Amount',hue='Gender',data = sm_age)
```

*From above both graphs , age between 26-35 have made more purchasing and are females*

## States

```python
df.columns
```

```python
sm_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
a=sns.barplot(x='State',y='Orders',data = sm_state)
for bars in a.containers:
    a.bar_label(bars)
```

```python
sm_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.barplot(x='State',y='Amount',data = sm_state)
```

From the above two graphs we can see that maximum orders and total sales/amounts are from Uttar Pradesh ,Maharashtra, Karnataka

# Marital Status

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
a=sns.countplot(data=df,x='Occupation')

for bars in a.containers:
    a.bar_label(bars)
```

```python
sm_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
a=sns.barplot(x='Occupation',y='Amount',data=sm_occ)
```

From the above graphs, maximum purchasing capacity is of IT Sector,Healthcare,Aviation sector

# Product Category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,5)})
a= sns.countplot(x='Product_Category',data=df)

for bars in a.containers:
    a.bar_label(bars)
```

```python
sm_pro=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.barplot(data=sm_pro,x='Product_Category',y='Amount')
sns.set(rc={'figure.figsize':(25,5)})
```

From the above graph it is clear that most selling products are Food,Clothing, Electronics

```python
sm_pro=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.barplot(data=sm_pro,x='Product_ID',y='Amount')
sns.set(rc={'figure.figsize':(25,5)})
```

# Conclusion 

Married women from age 26-35 years from Uttar Pradesh,Maharashtra,Karnataka working in IT Sector,Healthcare,Aviation sector
